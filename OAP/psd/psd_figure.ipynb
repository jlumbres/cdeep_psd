```python
%reload_ext autoreload
%autoreload 2
%matplotlib inline
#load modules
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from astropy import units as u
from astropy.io import fits
from  matplotlib.colors import LogNorm
from mpl_toolkits.axes_grid1 import make_axes_locatable
import pickle

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

from datetime import datetime
date_ran = datetime.now()
dr_str = date_ran.strftime("%d/%m/%Y %H:%M:%S")
print('Last date this notebook has been ran: {0}'.format(dr_str))
```

### Load Pickles
So many pickles

```python
def load_pickle(opt, sn_label):
    obj = []
    for n in range(0, len(sn_label)):
        filename='{0}/psd_radial_{0}_SN{1}.pickle'.format(opt, sn_label[n])
        with open(filename, 'rb') as data_file:
            pik_dict = pickle.load(data_file)
        rms_val = fits.open('{0}/psd_2D_{0}_SN{1}.fits'.format(opt, sn_label[n]))[0].header['rms_tot']
        pik_dict['rms_tot'] = rms_val*u.nm
        obj.append(pik_dict)
    return obj
```

```python
# MagAO-X OAP avg PSD
filename = 'psd_radial_MagAOX_oap_avg.pickle'
with open(filename, 'rb') as data_file:
    magaox = pickle.load(data_file)
magaox['rms_tot'] = 7.81331*u.nm # I forgot to add it, this is quickest
    
# open the other pickles
# M3 pickles
m3 = load_pickle(opt='M3', sn_label=[1,2])
oap1 = load_pickle(opt='OAP1', sn_label=[1,2])
oap2 = load_pickle(opt='OAP2', sn_label=[2,3])
oap3 = load_pickle(opt='OAP3', sn_label=[1,2])
```

### And now we plot

```python
obj = m3
sn_label = [1,2]
opt_label = 'M3'

matplotlib.rcParams["font.size"]=20
plt.figure(figsize=[16,10], dpi=100, facecolor='white')
for nf in range(0, len(sn_label)):
    plt.loglog(obj[nf]['k_radial'].value, 
               obj[nf]['psd_radial'].value, 
               linewidth=2.5,
               label='SN{0}\nRMS={1:.3f}'.format(sn_label[nf], obj[nf]['rms_tot']))
plt.loglog(magaox['k_radial'].value, magaox['psd_radial'].value, linewidth=2.5, color='k',
           label='MagAO-X avg oap\nRMS={0:.3f}'.format(magaox['rms_tot']))
plt.xlabel('Spatial Frequency [{0}]'.format(r'$m^{-1}$'))
plt.ylabel('Power Spectrum [{0}{1}]'.format(r'$nm^{2}$',r'$m^{2}$'))
plt.legend(prop={'size':14})#,loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('{0} PSD'.format(opt_label))
```
